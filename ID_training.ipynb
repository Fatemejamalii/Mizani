```python
import pandas as pd
import os
import shutil
number = -1
df = pd.DataFrame(columns=['filename', 'newname'])
for k in range(0,50000,2000):
  base_foldername  = f'{k:05d}'
  for filename in os.listdir('/content/drive/MyDrive/mizani/Dataset_256/dataset_256_1/images/'+ base_foldername + '/'):
    if os.path.isfile('/content/drive/MyDrive/mizani/Dataset_256/dataset_256_1/ws/' + base_foldername + '/' + filename[0:-4] + '.npy') and \
       os.path.isfile('/content/drive/MyDrive/mizani/Dataset_256/image_masks/'  + filename):   

      number +=1
      newname = f'{number:05d}'
      new_foldername = f'{int(number - number % 1e3):05d}'

      shutil.move('/content/drive/MyDrive/mizani/Dataset_256/dataset_256_1/images/'+ base_foldername + '/' + filename,
                  '/content/drive/MyDrive/mizani/Dataset_256/dataset_256/images/'+ new_foldername +'/' + newname + '.png' )
      
      shutil.move('/content/drive/MyDrive/mizani/Dataset_256/dataset_256_1/ws/'+ base_foldername + '/' + filename[0:-4]  + '.npy',
                  '/content/drive/MyDrive/mizani/Dataset_256/dataset_256/ws/'+ new_foldername +'/' + newname + '.npy' )
      
      shutil.move('/content/drive/MyDrive/mizani/Dataset_256/image_masks/' + filename,
                  '/content/drive/MyDrive/mizani/Dataset_256/dataset_256/image_masks/'+ new_foldername +'/' + newname + '.png')
      
      df =df.append({'filename':filename, 'newname':newname +'.png'}, ignore_index=True)
      
      print(base_foldername,filename, new_foldername ,newname )
      
```
